```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('../Projektdatensaetze/2015-2017/501_2015bis2017KennedybrueckeNordseite.csv',
                 sep=';')
df = df.rename(columns={df.columns[0]: 'Datum'})
df['Datum'] = pd.to_datetime(df['Datum'])
df['Datum'] = df['Datum'].dt.strftime('%Y/%m')
df
```

```python
df.plot(x=df.columns[0], y=df.columns[1], color='blue', ylabel='Anzahl', xlabel='Datum', title=df.columns[1], label='Anzahl Fahrradfahrer')
```
